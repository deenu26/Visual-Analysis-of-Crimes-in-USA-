```python
cityvscrime = pd.read_excel('city_vs_crime2.xlsx')
df_cityvscrime = pd.DataFrame(cityvscrime)
df_cityvscrime = df_cityvscrime.fillna(0)
df_cityvscrime = df_cityvscrime.set_index(['City'])
df_cityvscrime
```

```python
plt.scatter(df_cityvscrime['Theft From Motor Vehicle'],df_cityvscrime['Theft of Motor Vehicle Parts or Accessories'])
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn import preprocessing
df_cityvscrime_standardized = preprocessing.scale(df_cityvscrime)
print(df_cityvscrime_standardized)
df_cityvscrime_standardized = pd.DataFrame(df_cityvscrime_standardized)
```

```python
plt.figure(figsize=(10, 8))
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df_cityvscrime_standardized)
    wcss.append(kmeans.inertia_) #criterion based on which K-means clustering works
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(df_cityvscrime_standardized)

y_kmeans
```

```python
#beginning of  the cluster numbering with 1 instead of 0
y_kmeans1=y_kmeans+1

# New list called cluster
cluster = list(y_kmeans1)
# Adding cluster to our data set
df_cityvscrime['cluster'] = cluster
```

```python
#Mean of clusters 1 to 4
kmeans_mean_cluster = pd.DataFrame(round(df_cityvscrime.groupby('cluster').mean(),1))
kmeans_mean_cluster
```

```python
import seaborn as sns

plt.figure(figsize=(12,6))
sns.scatterplot(x=df_cityvscrime['Theft From Motor Vehicle'], y = df_cityvscrime['Theft of Motor Vehicle Parts or Accessories'],hue=y_kmeans1)
```
